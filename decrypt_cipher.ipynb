```python
def rail_fence_decrypt(ciphertext, key):
    rail_fence = [[''] * len(ciphertext) for _ in range(key)]
    direction = 1  # 1 represents moving down, -1 represents moving up
    row = 0

    for char in ciphertext:
        rail_fence[row][0] = 'x'  # Placeholder for the rail boundary
        rail_fence[row][-1] = 'x'

        rail_fence[row][0] = char
        row += direction

        if row == key - 1 or row == 0:
            direction *= -1

    plaintext = ''
    for row in rail_fence:
        for char in row:
            if char.isalpha():
                plaintext += char

    return plaintext

ciphertext = "I FOSW TUK NSIT IELR TA GNLG LAHSNELMIM"
key = 5

plaintext = rail_fence_decrypt(ciphertext, key)
print(plaintext)
```
